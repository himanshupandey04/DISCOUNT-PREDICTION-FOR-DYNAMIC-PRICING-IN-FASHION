```python
##Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Category'] = le.fit_transform(df['Category'])
df['Brand'] = le.fit_transform(df['Brand'])
df['Season'] = le.fit_transform(df['Season'])
```

```python
##Define Features and Target
X = df.drop(columns=['Discount', 'Product_ID'])  # Drop ID and target
y = df['Discount']
```

```python
##Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
##Scale Features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
##Split into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
##Train Linear Regression Model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
##Evaluate Linear Regression Model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred_lr)
mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_lr)

print("Linear Regression Evaluation:")
print(f"MAE : {mae:.2f}")
print(f"MSE : {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R   : {r2:.4f}")
```

```python
##Ridge & Lasso Regression
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge()
ridge.fit(X_train, y_train)
print("Ridge R:", r2_score(y_test, ridge.predict(X_test)))

lasso = Lasso()
lasso.fit(X_train, y_train)
print("Lasso R:", r2_score(y_test, lasso.predict(X_test)))
```

```python
##Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
print("Decision Tree R:", r2_score(y_test, dt.predict(X_test)))
```

```python
##Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
print("Random Forest R:", r2_score(y_test, rf.predict(X_test)))
```
